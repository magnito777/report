```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your DataFrame from a CSV file
df = pd.read_csv('bonelab - tibias.csv')

count = df['oaota-region'].value_counts()
roles = df['roles'].value_counts()
# display count and percentage of this column oaota-region
print(df['oaota-region'].value_counts())
print(df['oaota-region'].value_counts(normalize=True) * 100)

# visualise bar chart roles should be colored in bars
plt.figure(figsize=(10, 6))
df_grouped = df.groupby(['oaota-region', 'roles']).size().unstack(fill_value=0)
ax = df_grouped.plot(kind='bar', stacked=True, figsize=(10, 6), color=sns.color_palette("husl", len(df['roles'].unique())))
plt.title('Distribution of oaota-region by Roles')
plt.xlabel('oaota-region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Roles')

# Add numbers on bars
for container in ax.containers:
    ax.bar_label(container, label_type='center')

plt.tight_layout()
plt.savefig('oaota-region_roles_distribution_stacked.png')
plt.show()


```

```python
# cross tabulate the two columns
df_cross = pd.crosstab(df['frcture-type'], df['roles'])
fracture_table = pd.crosstab(df['frcture-type'], df['roles'])
fracture_table['All Patients'] = fracture_table.sum(axis=1)
# Add percentage to the table
fracture_table_percentage = fracture_table.div(fracture_table['All Patients'], axis=0) * 100
fracture_table_percentage = fracture_table_percentage.round(2)
combined_table_str = fracture_table.astype(str) + " (" + fracture_table_percentage.round(1).astype(str) + "%)"
# Display the table
display(combined_table_str)

```
